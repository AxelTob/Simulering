```python
import pandas as pd
import matplotlib.pyplot as plt
```

## EITN95: Sumulation - Assignment 1 - Analysis/Plots for Task1

### Import Data

```python
df_11 = pd.read_csv('Customers_in_Queue_1.1.csv')
df_11.head()
```

```python
df_15 = pd.read_csv('Customers_in_Queue_1.5.csv')
df_15.head()
```

```python
df_20 = pd.read_csv('Customers_in_Queue_2.0.csv')
df_20.head()
```

```python
df_11 = df_11.dropna()
df_15 = df_15.dropna()
df_20 = df_20.dropna()
```

```python
df_list = [df_11, df_15, df_20]
```

```python
param_df = [1.1,1.5,2]
```

### Boxplots for number of customers in the queues

```python
fig = plt.figure(figsize =(6, 4))
#ax = fig.add_axes([0, 0, 1, 1])
# Creating plot
plt.boxplot([df_11['Queue1'], df_15['Queue1'], df_20['Queue1']],showfliers=False, showmeans=True)
plt.xticks([1, 2, 3], param_df)
plt.xlabel('Time between arrivals [s]')
plt.ylabel('Customers in Queue')
#plt.ylim([0,40])
#plt.title('Queue 1')
plt.savefig('Task3_Queue1.png')
```

```python
fig = plt.figure(figsize =(6, 4))
plt.boxplot([df_11['Queue2'], df_15['Queue2'], df_20['Queue2']],showfliers=False, showmeans=True)
plt.xticks([1, 2, 3], param_df)
plt.xlabel('Time between arrivals [s]')
plt.ylabel('Customers in Queue')
#plt.title('Queue 2')
#plt.ylim([0,40])
plt.savefig('Task3_Queue2.png')
```

### Graphs for number of customer in the queues for first 2,000 s

```python
for i in range(0,len(df_list)):
    fig = plt.figure()
    fig = plt.figure(figsize =(6, 4))
    df = df_list[i]
    param = param_df[i]
    fid = plt.figure()
    df = df[df['Time']<2000]
    plt.plot(df['Time'], df['Queue1'], label = 'Queue 1')
    plt.plot(df['Time'], df['Queue2'], label = 'Queue 2')
    plt.legend(loc = 'upper left')
    plt.ylabel('# Customers')
    plt.xlabel('Time [s]')
    plt.title('Time between arrivals: ' + str(param) + " s")
    plt.savefig('Task3_Arrivals' + str(int(param*10)) + '.png')
```
